```python
import numpy as np

vtkpath = r'/Users/msb/Desktop/Step-Compression_100_with_data.vtu'
vtkpv = pv.read(vtkpath)

# clip dataset
clip = vtkpv.clip(normal='y', invert=False)

# Determine the range of 'IsovolumicDeformation' you want to focus on
# For example, if you know your values of interest lie between 0.1 and the maximum value in the dataset
min_value = 1.5
max_value = clip['IsovolumicDeformation'].max()

# Adjust the scalar range to enhance contrast on higher values
scalar_range = (min_value, max_value)

# plot IsovolumicDeformation from vtkpv celldata with adjusted scalar range
p = pv.Plotter()
p.add_mesh(clip, scalars='IsovolumicDeformation', show_edges=False, cmap='viridis', log_scale=True, scalar_bar_args={"title": "IsovolumicDeformation", "n_labels": 5}, clim=scalar_range)
p.view_xz()
p.camera.roll += 180
p.show()
```
